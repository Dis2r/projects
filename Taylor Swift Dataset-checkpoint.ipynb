Добавить столбец с длительностью в секундах 

```python
def ms_to_s(row):
    duration_seconds = row('duration_ms')/1000
    duration_seconds = duration_seconds.to_frame()
    return duration_seconds
songs['duration_s'] = songs.apply(ms_to_s, axis=1)

#songs.insert(-1, 'duration_min', songs['duration_ms']/1000)
```
